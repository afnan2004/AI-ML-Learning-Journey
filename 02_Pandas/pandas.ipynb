```python
import pandas as pd

students = pd.DataFrame({
    'id':   [1, 2, 3, 4],
    'name': ['Arjun', 'Priya', 'Ravi', 'Meena']
})
scores = pd.DataFrame({
    'id':    [1, 2, 3],
    'score': [85, 92, 78]
})

# Inner — Meena (id=4) is DROPPED, no matching score
df=pd.merge(students, scores, on='id', how='inner')
df
```

```python
# Left — Meena is KEPT, score becomes NaN
df=pd.merge(students, scores, on='id', how='left')
df
```
